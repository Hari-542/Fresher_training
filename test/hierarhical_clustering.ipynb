```python
import pandas as pd
```

```python
data = pd.read_csv('Live.csv')
data.isnull().sum(),data.head()

datasubset = data.loc[:,["num_reactions","num_comments"]]
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
datasubset = ss.fit_transform(datasubset)

```

```python
import scipy.cluster.hierarchy as sch

dendogram = sch.dendrogram(sch.linkage(datasubset,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering

ac =  AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
cluster = ac.fit_predict(datasubset)
cluster
```

```python
from sklearn.metrics import silhouette_score

ss = silhouette_score(datasubset,cluster)
ss
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=datasubset[:,0],y=datasubset[:,1],c=cluster)
```
