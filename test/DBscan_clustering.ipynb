```python
import pandas as pd
```

```python
df = pd.read_csv('Live.csv')

data = df.loc[:,["num_reactions","num_comments"]]

react = data.num_reactions
comment = data.num_comments
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=react,y=comment)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
x = ss.fit_transform(data)

x
```

```python
from sklearn.cluster import DBSCAN 

model = DBSCAN(eps=0.19,min_samples=50).fit_predict(x)
model
```

```python
plt.scatter(x=x[:,0],y=x[:,1],c=model)
```

```python
from sklearn.metrics import silhouette_score as ss

ss(x,model)
```
